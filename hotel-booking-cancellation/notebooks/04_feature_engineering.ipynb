```python
import pandas as pd

df = pd.read_csv("../datasets/hotel_bookings_clean.csv")
df.shape
```

```python
# Noches totales de estancia
df["total_nights"] = df["stays_in_weekend_nights"] + df["stays_in_week_nights"]
```

```python
# Variable binaria de si la reserva incluye niños o bebés
df["is_family"] = ((df["children"] > 0) | (df["babies"] > 0)).astype(int)
```

```python
# Mes de llegada como número
meses = {
    "January": 1, "February": 2, "March": 3, "April": 4,
    "May": 5, "June": 6, "July": 7, "August": 8,
    "September": 9, "October": 10, "November": 11, "December": 12
}
df["arrival_month_num"] = df["arrival_date_month"].map(meses)
```

```python
df.groupby("is_repeated_guest")["is_canceled"].mean()
```

```python
print(df.groupby("is_family")["is_canceled"].mean())
print(df.groupby("arrival_month_num")["is_canceled"].mean())
```

```python
df.groupby("total_of_special_requests")["is_canceled"].mean()
```

```python
df.to_csv("../datasets/hotel_bookings_features.csv", index=False)
```
